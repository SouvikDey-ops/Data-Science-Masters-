```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
```

```python
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df['deperture_hour']=df['Dep_Time'].str.split(':').str[0]
df['deperture_minute']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['deperture_hour']=df['deperture_hour'].astype(int)
df['deperture_minute']=df['deperture_minute'].astype(int)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
df['arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['arrival_minute']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df['arrival_hour']=df['arrival_hour'].astype(int)
df['arrival_minute']=df['arrival_minute'].astype(int)
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['duration_in_hour']=df['Duration'].apply(lambda x: x.split()[0])
```

```python
df['duration_in_minute']=df['Duration'].apply(lambda x: x.split()[0])
```

```python
df.head(2)
```

```python
df.drop('duration_in_minute', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'].fillna(df['Total_Stops'].mode()[0], inplace=True)
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4})
```

```python
df.head(10)
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df
```

```python
df['Duration'].str.split(' ').str[1]#.str.split('m')[0]
```

```python
df['Duration_in_hours']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration_in_minutes']=df['Duration'].str.split(' ').str[1].fillna('0m').str.split('m').str[0]
```

```python
df.head(2)
```

```python
df['Duration_in_hours']=df['Duration_in_hours'].apply(pd.to_numeric, errors='coerce')
df['Duration_in_minutes']=df['Duration_in_minutes'].apply(pd.to_numeric, errors='coerce')
```

```python
df['Duration(minutes)'] = df['Duration_in_hours']*60 + df['Duration_in_minutes']
```

```python
df.drop('Duration', axis=1, inplace=True)
```
