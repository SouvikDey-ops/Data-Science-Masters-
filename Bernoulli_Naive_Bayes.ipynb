## Bernoulli Naive Byes
Predicting 'Time' from the 'tips' dataset

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
# Encoding the categirical features with OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['sex', 'smoker', 'day', 'size']]).toarray()
```

```python
encoded
```

```python
encoded_data = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoded_data
```

```python
# Encoding the output feature 'time' with LabelEncoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['time'] = encoder.fit_transform(df['time'])
```

```python
df.head()
```

```python
# Concating the dataframes- df and encoded_data
Data = pd.concat([df, encoded_data], axis=1)
```

```python
Data
```

```python
# Dropping the categorical features
Data = Data.drop(['sex', 'smoker', 'day', 'size'], axis=1)
```

```python
Data
```

```python
# Independent data(x) and dependent data(y)
x = Data.drop('time',axis=1)
y = Data['time']
```

```python
x
```

```python
y
```

```python
# Splitting into training and test data
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
# Model Traning
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
y_pred = bnb.predict(x_test)
```

```python
y_pred
```

```python
# Checking model accuracy
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print('The model accuracy is', accuracy_score(y_test, y_pred)*100, '%')
print('Classification Report --')
print(classification_report(y_test, y_pred))
print('Confusion Matrix --')
print(confusion_matrix(y_test, y_pred))
```
