## Decison Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
data = load_iris()
```

```python
print(data.DESCR)
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
x=df.drop('species', axis=1)
```

```python
y = data.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decisiontree = DecisionTreeClassifier()
```

```python
decisiontree.fit(x_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,6))
tree.plot_tree(decisiontree, filled=True)
```

```python
# Post-Prunning: After the second level the further splitting was not required. SP max-depth is chones to be equal to 2
decisiontree = DecisionTreeClassifier(max_depth=2)
decisiontree.fit(x_train, y_train)
```

```python
plt.figure(figsize=(10,6))
tree.plot_tree(decisiontree, filled=True)
```

```python
# Precition
y_pred=decisiontree.predict(x_test)
```

```python
y_pred
```

```python
# Checking the model accuracy 
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_test, y_pred)
print(score)
report = classification_report(y_test, y_pred)
print(report)
```

```python
# Pre-Prunning with hyperparameter tuning
import warnings
warnings.filterwarnings('ignore')

parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
}

from sklearn.model_selection import GridSearchCV

decisiontree = DecisionTreeClassifier()
clf=GridSearchCV(decisiontree, param_grid=parameter, cv=5, scoring='accuracy')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.best_params_
```

```python
# Now we would prepare out model using the above parameters

decisiontree = DecisionTreeClassifier(criterion= 'gini',
 max_depth= 4,
 max_features= 'auto',
 splitter= 'random')

decisiontree.fit(x_train, y_train)
y_pred = decisiontree.predict(x_test)
```

```python
y_pred
```

```python
# Checking the model accuracy 
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_test, y_pred)
print(score)
report = classification_report(y_test, y_pred)
print(report)
```
