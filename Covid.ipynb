# Analysis of Covid Data

```python
import pandas as pd
```

```python
df = pd.read_csv('covid_data.csv', encoding='latin-1')
```

```python
df.head()
```

# Cleaning of the Dataset

## There are certain records in the dataset where the country is same as the continent. So we must clean the data by dropping those records.

```python
df.drop(df[df['Country'] == 'Europe'].index, inplace = True)
```

```python
df.drop(df[df['Country'] == 'Asia'].index, inplace = True)
```

```python
df.drop(df[df['Country'] == 'Africa'].index, inplace = True)
```

```python
df.drop(df[df['Country'] == 'North-America'].index, inplace = True)
```

```python
df.drop(df[df['Country'] == 'South-America'].index, inplace = True)
```

```python
df.drop(df[df['Country'] == 'Oceania'].index, inplace = True)
```

```python
df.drop(df[df['Country'] == 'All'].index, inplace = True)
```

```python
df
```

## So it can be seen the China is the most populous country in the world.

# Which is the least populous country?

# What percentage of the country's population was infected?

```python
df['Percentage_Infected = (Cases.Total/Population)*100']=(df['Cases.Total']/df['Population'])*100
```

```python
df.head()
```

# Rank the countries on the basis of percentage infected and find out which country has the most percentage of infection.

```python
df1 = df.sort_values('Percentage_Infected = (Cases.Total/Population)*100', ascending = False)
```

```python
df1.head()
```

```python
Data_percent_infected = df1[['Country']]
```

```python
Data_percent_infected.head()
```

```python
Data_percent_infected['Rank(basis: percent_infected)'] = [i for i in range(1,len(Data_percent_infected)+1)]
```

```python
Data_percent_infected.head()
```

```python
Ranked_Data = df.merge(Data_percent_infected, on='Country')
```

```python
Ranked_Data.head(10)
```

## The above dataframe (Ranked_Data) shows the rank of each country on the basis of percentage of the population infected.

```python
Ranked_Data[Ranked_Data['Rank(basis: percent_infected)']==1]
```

## So it can be seen that San-Marino has the highest infection percentage.

# Rank the countries on the basis of total cases (decreasing order) and find out which country has the highest total infection cases.

```python
df2 = df.sort_values('Cases.Total', ascending = False)
```

```python
df2.head()
```

```python
Data_total_cases = df2[['Country']]
```

```python
Data_total_cases.head()
```

```python
Data_total_cases['Rank(basis: total_cases)'] = [i for i in range(1,len(Data_total_cases)+1)]
```

```python
Data_total_cases.head()
```

```python
Ranked_Data_2 = Ranked_Data.merge(Data_total_cases, on='Country')
```

```python
Ranked_Data_2.head(10)
```

## The above dataframe (Ranked_Data_2) shows the rank of each country on the basis of total cases of infection.

```python
Ranked_Data_2[Ranked_Data_2['Rank(basis: total_cases)']==1]
```

## So it can be seen that USA has the most number of total cases of infection.

```python
Data_fatality_rate = df3[['Country']]
```

## The above dataframe (Ranked_Data_3) shows the rank of each country on the basis of fatality rate.

# Rank the countries on the basis of recovery rate (i.e. percentage of the total infected people recovered) and find out which country has the highest recovery rate - in decreasing order

```python
Ranked_Data_3['Recovery_Rate = (Cases.Recovered)/Cases.Total)*100']=(df['Cases.Recovered']/df['Cases.Total'])*100
```

```python
Data_recovery_rate.head()
```

```python
Data_recovery_rate.head()
```

```python
Ranked_Data_4 = Ranked_Data_3.merge(Data_recovery_rate, on='Country')
```

```python
Ranked_Data_4.head()
```

## The above dataframe shows the rank of the countries on the basis of the recovery rate.

```python
Ranked_Data_4[Ranked_Data_4['Rank(basis: recovery_rate)']==1]
```
