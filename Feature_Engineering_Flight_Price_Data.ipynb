## Feature Engineering

#### Using Flight Price data

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel('flight_price.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
# Splitting the 'Date_of_Journry' into date, month, and year 
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
# Changing the datetype of Date, Month, and Year into integer
df['Date']=df['Date'].astype(int)
df['Month']=df['Date'].astype(int)
df['Year']=df['Date'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Dropping the column 'Route' as Source and Destination are already present in the dataset
df.drop('Route', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Checking missing values 
df.isnull().sum()
```

```python
mode_value=df[df['Total_Stops'].notna()]['Total_Stops'].mode()[0]
```

```python
df['Total_Stops']=df['Total_Stops'].fillna(mode_value)
```

```python
df.isnull().sum()
```

```python
df['Arrival_Time']
```

```python
# Removing all unwanted informations from the 'Arrival_Time' column
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head()
```

```python
df['Dept_hr']=df['Dep_Time'].str.split(':').str[0]
df['Dept_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Arriv_hr']=df['Arrival_Time'].str.split(':').str[0]
df['Arriv_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df[['Dept_hr',	'Dept_min', 'Arriv_hr',	'Arriv_min']]=df[['Dept_hr',	'Dept_min', 'Arriv_hr',	'Arriv_min']].astype(int)
```

```python
df.info()
```

```python
df.drop(df[['Dep_Time',	'Arrival_Time']], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Converting the 'Duration' column to minutes

def convertion(duration):
    minutes = 0
    
    time_parts = duration.split()
    
    for part in time_parts:
        if 'h' in part:
            hours = int(part.replace('h','')) * 60
            minutes += hours
        
        elif 'm' in part:
            minutes += int(part.replace('m',''))
            
    return minutes
```

```python
df['Duration_in_min']=df['Duration'].apply(convertion)
```

```python
df.head()
```

```python
df.drop('Duration', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Labelling the 'Total_Stops' column 
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4})
```

```python
df.head()
```

```python
df.info()
```

```python
# Encoding the categorical columns 'Airline', 'Source', 'Destination', 'Additional_Info' using OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[['Airline', 'Source', 'Destination', 'Additional_Info']]).toarray()
```

```python
code = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
df=pd.concat([df, code], axis=1)
```
